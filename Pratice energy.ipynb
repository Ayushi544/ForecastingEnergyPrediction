```python
import pandas as pd
df_Energy = pd.read_csv("A:/2024/CDAC/December 2024/Project Energy/Electricity+Consumption/powerconsumption.csv")
df_Energy = df_Energy.set_index('Datetime')
df_Energy.index = pd.to_datetime(df_Energy.index)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#print(type(df_Energy))
color_pal = sns.color_palette('Set2', n_colors=len(df_Energy.columns))
df_Energy.plot(style='-', figsize=(15, 5), color=color_pal, title='Energy consumption in zone 1')
```

```python
df_Energy['PowerConsumption_Zone1'].plot(
    style='-', 
    figsize=(15, 5), 
    color=color_pal[0],  # Select the first color
    title='Energy Consumption in Zone 1'
)

# Add labels
plt.xlabel('Datetime')
plt.ylabel('Power Consumption (kWh)')

# Show the plot
plt.show()
```

```python
train = df_Energy.loc['July 2014':'December 2016']
test= df_Energy.loc['January 2017':'December 2017']
fig, ax = plt.subplots(figsize=(15, 5))
train.plot(ax=ax, label='Training Set')
test.plot(ax=ax, label='Test Set')
ax.axvline('2017-09-31', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
plt.show()
```
